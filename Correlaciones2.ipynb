```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv sin indice
Credicel = pd.read_excel('PrepCrit3.xlsx')
```

```python
Credicel.info()
```

```python
#Corroboramos valores nulos
ValoresNulos = Credicel.isnull().sum()
ValoresNulos
```

```python
#Filtro por objeto (Primer Filtro)
Credicel26Sem = Credicel[Credicel["plazo"] == "26S"]
Credicel26Sem
```

### 26 Semanas
Correlación y modelo

```python
# Scatter plot entre la variable dependiente (total) e independientes
# Comportamiento de su dispersión

from turtle import color

sns.scatterplot(x = 'enganche', y = 'precio', color = 'blue', data = Credicel26Sem)
```

```python
# Variable Dependiente e Independiente

Var_Indep = Credicel26Sem[['enganche']]
Var_Dep = Credicel26Sem['precio']
```

```python
# Se define model como la función de regresión lineal

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Verificamos la función relacionada al modelo
type(model)
```

```python
# Ajustamos el modelo con las variables declaradas
model.fit(X = Var_Indep, y = Var_Dep)
# X debe ser mayúscula para la variable independiente
```

```python
# Verificamos los coeficientes obtenidos para el modelo ajustado

model.__dict__
```

Modelo Matemático (Precio vs Enganche): y = 2.39642446x1 + 1822.188552100767

```python
# Corroboramos cual es el coeficiente de Determinación de nuestro modelo

coef_Deter = model.score(X = Var_Indep, y = Var_Dep)
coef_Deter
```

```python
# Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl = np.sqrt(coef_Deter)
coef_Correl
```
